```python
from keras.preprocessing.image import ImageDataGenerator
from skimage import io
import numpy as np
import os
from PIL import Image
```

```python
datagen = ImageDataGenerator(        
        rotation_range = 40,
        shear_range = 0.2,
        zoom_range = 0.2,
        horizontal_flip = True,
        brightness_range = (0.5, 1.5))

SIZE = 256
```

```python
datagen = ImageDataGenerator(        
        zoom_range = 0.5,
        brightness_range = (0.5, 1.5))

SIZE = 256
```

## 1. Bacterial Leaf Blight

```python
blb = '/content/drive/MyDrive/SEM 6 - MBKM/PROYEK AKHIR/DATASET/Padi/train_images/bacterial_leaf_blight/'
blb_data = []
blb_images = os.listdir(blb)

for i, image_name in enumerate(blb_images):    
    if (image_name.split('.')[1] == 'jpg'):        
        image = io.imread(blb + image_name)        
        image = Image.fromarray(image, 'RGB')        
        image = image.resize((SIZE,SIZE)) 
        blb_data.append(np.array(image))

x = np.array(blb_data)
i = 0

for batch in datagen.flow(x, batch_size=16,
                          save_to_dir='/content/drive/MyDrive/SEM 6 - MBKM/PROYEK AKHIR/DATASET/Padi/Augmentasi/Blb',
                          save_prefix='aug',
                          save_format='jpg'):    
    i += 1    
    if i > 10:        
        break
```

```python
aug_dir = '/content/drive/MyDrive/SEM 6 - MBKM/PROYEK AKHIR/DATASET/Padi/Augmentasi/Blb'
main_dir = '/content/drive/MyDrive/SEM 6 - MBKM/PROYEK AKHIR/DATASET/Padi/train_images/bacterial_leaf_blight'
labels_aug = os.listdir(aug_dir)
labels_main = os.listdir(main_dir)
p1 = len(labels_aug) + len(labels_main)
```
